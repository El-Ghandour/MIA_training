```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
df =pd.read_csv("Weather Dataset.csv")
import csv
```

```python
df =df.drop_duplicates()
```

```python
df = df[df['Temperature (C)'] != '']
df = df[df['Temperature (C)'] <50]
df = df[df['Temperature (C)'] >-30]
df = df[df['Humidity'] != '']
df = df[df['Humidity'] >-1]
df = df[df['Apparent Temperature (C)'] != '']
df = df[df['Apparent Temperature (C)'] <50]
df = df[df['Apparent Temperature (C)'] > -30]
df = df[df["Pressure (millibars)"]>1000]
df = df[df["Wind Bearing (degrees)"] >20]
#df = df[df["Loud Cover"] >= 0]
```

```python

df.plot(x='Formatted Date', y='Temperature (C)') 
plt.xticks(visible = False)
plt.show()
```

```python
df.plot(kind='scatter',x='Humidity',y='Temperature (C)')
#plt.scatter(x='Humidity', y='Temperature (C)')
plt.show()
```
